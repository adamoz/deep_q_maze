```python
import logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)
```

```python
from qrl_navigation.agent import Agent, WeightedAgent
from qrl_navigation.environment import BananaMazeEnv
from qrl_navigation.training import dqn_training
import torch
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
maze_env = BananaMazeEnv(train_mode=True)
agent = WeightedAgent(maze_env.state_size, maze_env.action_size, device=device, dueling=True)
```

```python
meta = dqn_training(agent, maze_env, '../experiments/dueling_weighted/', logger=logger, n_episodes=1000)
```

```python
maze_env.close()
```
